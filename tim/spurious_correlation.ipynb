```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# ノイズの生成
num_data = 200
np.random.seed(1)
e_x = np.random.normal(size=num_data)
e_y = np.random.normal(size=num_data)

# データ生成
Y = e_y
X = 2*Y + e_x

# 相関係数を計算
round(np.corrcoef(X, Y)[0,1], 3)
```

```python
plt.figure(figsize=[6,4.5], dpi=200)
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# ノイズの生成
num_data = 200
np.random.seed(2)
e_x = np.random.normal(size=num_data)
e_y = np.random.normal(size=num_data)
e_z = np.random.normal(size=num_data)

# データ生成
Z = e_z
X = 2*Z + e_x
Y = 3*Z + e_y

# 相関係数を計算
round(np.corrcoef(X, Y)[0,1], 3)
```

```python
plt.figure(figsize=[6,4.5], dpi=200)
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# ノイズの生成
num_data = 200
np.random.seed(3)
e_x = np.random.normal(size=num_data)
e_y = np.random.normal(size=num_data)

# データ生成
X = e_x
Y = e_y

# 合流点を作成
Z = X + Y

# Zについての条件でデータを選抜
mask = Z > 1
X_new = X[mask]
Y_new = Y[mask]

# 相関係数を計算
round(np.corrcoef(X_new, Y_new)[0,1], 3)
```

```python
plt.figure(figsize=[6,4.5], dpi=200)
plt.scatter(X_new, Y_new)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
